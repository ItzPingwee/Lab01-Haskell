```haskell
cuadrado :: Int -> Int
cuadrado x = x * x

suma :: Int -> Int -> Int
suma x y = x + y

sumaCuadrados :: Int -> Int -> Int
sumaCuadrados a b = suma (cuadrado a) (cuadrado b)
```

```haskell
sumaCuadrados 2 2
```

```haskell
duplicar :: Int -> Int
duplicar x = x * 2

dobleLista :: [Int] -> [Int]
dobleLista lista = map duplicar lista
```

```haskell
dobleLista[2, 4, 6]
```

```haskell
-- Una forma mas simple de haber hecho este proceso es mediante el calculo lambda, que no requiere definir la funcion duplicar, seria algo como dobleLista lista = map(*2) lista.
```

```haskell
clasificarNota :: Int -> String
clasificarNota n
    | n < 3          = "Reprobado"
    | n >= 3 && n < 4 = "En recuperación"
    | n >= 4          = "Aprobado"
    | otherwise       = "Nota no válida"
```

```haskell
clasificarNota 4
```

```haskell
fibonacci :: Int -> Int
fibonacci 0 = 0
fibonacci 1 = 1
fibonacci n = fibonacci (n - 1) + fibonacci (n - 2)
```

```haskell
fibonacci 6
```
