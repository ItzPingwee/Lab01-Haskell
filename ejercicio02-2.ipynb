```haskell
-- Se relaciona con el lazy evaluation ya que Haskell es no calcula nada hasta que realmente lo necesita. Si se escribe [1..1000000], Haskell no crea esa lista gigante de inmediato.
-- Solo va generando los números a medida que otras funciones se los piden.
```

```haskell
primerElemento :: [a] -> a
primerElemento (x:_) = x
primerElemento (x:xs) = x
primerElemento [] = error "Lista vacia"
primerElemento [1, 2, 3]
```

```haskell
productoLista :: [Int] -> Int
productoLista [] = 1
productoLista (x:xs) = x * productoLista xs
productoLista[1, 2, 3, 4]
```

```haskell
eliminarNegativos :: [Int] -> [Int]
eliminarNegativos [] = []
eliminarNegativos (x:xs)
    | x >= 0    = x : eliminarNegativos xs
    | otherwise = eliminarNegativos xs
```

```haskell
eliminarNegativos [-1, 2, -3, 5]
```

```haskell
invertir :: [a] -> [a]
invertir [] = []
invertir (x:xs) = invertir xs ++ [x]
```

```haskell
invertir [1, 2, 3]
```
